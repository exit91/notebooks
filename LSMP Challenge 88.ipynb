> Let $x_1 = 1$ and $nx_n^2 - x_{n-1} - n = 0$ for $n \geq 2$.
> Find $\lim_{n\to \infty} x_n$ or show that it diverges.

We show that $x_n$ converges, if we always take the positive solution $x_n = \sqrt{1 + \frac{x_{n-1}}{n}}$.

With some extra work it would then follow that $x_n$ is in general divergent if we can switch signs.

A useful inequality derived from the convexity of the square root function:

$$ \sqrt{1+x} \leq 1 + \frac{x}{2} \quad \text{ for } x \geq -1 $$

```python
var('x')
plot(sqrt(1+x),-1,4) + plot(1 + x/2,-1,4,color="red")
```

Concavity also yields
$$ 1+x \leq \sqrt{1+x} \text{ for } -1 \leq x \leq 0 $$

```python
plot(sqrt(1+x),-1,0) + plot(1+x,-1,0,color="red")
```

Claim: $||x_n| - 1| \leq \frac{2}{n}$.

For $n = 1$ this is evidently true.
Assume $n \geq 2$ and $||x_{n-1}| - 1| \leq \frac{2}{n}$.

1. $-2 \leq x_{n-1} \leq 2$
2. If $x_{n-1} \geq 0$, $||x_n| - 1 | = \sqrt{1 + \frac{x_{n-1}}{n}} - 1 \leq \frac{x_{n-1}}{2n} \leq \frac{1}{n}$
3. If $x_{n-1} < 0$, $||x_n| - 1 |
       = 1 - \sqrt{1 + \frac{x_{n-1}}{n}}
       \leq 1 - \sqrt{1 - \frac{2}{n}}
       \leq 1 - (1 - \frac{2}{n}) = \frac{2}{n}$.
